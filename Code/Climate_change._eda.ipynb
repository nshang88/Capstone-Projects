```python
# Calculate the yearly average temperature of each city
climate_yearly=climate_data.groupby(pd.DatetimeIndex(climate_data.index).year).mean()
climate_yearly
```

```python
sns.pairplot(climate_yearly, vars=['AK_Anchorage_Temp','Milwaukee_Temp'], kind ='scatter' );
```

```python
climate_yearly.describe()
```

```python
city_data=[]
```

```python
for city in climate_yearly.columns:
    city_data.append({city:[climate_yearly[city].mean(), climate_yearly[city].std()]})
```

```python
city_data
```

```python
#assume there is no temperature change, the yearly mean temperature follows standard distribution, 
#generate 10000 yearly mean temperature for each city

city_random=[]
for x in range(20):
    y=list(city_data[x].values())
    city_random.append([np.random.normal(y[0][0], y[0][1], 10000)])
city_random[1]
```

```python
#Generate one data for the difference between 5 years mean temperature of 20 cities

b=[]
for x in range(20):
        array=np.array(city_random[x])
        LIST=array.tolist()[0]
        a1=np.mean(np.random.choice(LIST, 5))
        a2=np.mean(np.random.choice(LIST, 5))
        b.append(a2-a1)
print(b)
print(np.mean(b))
```

```python
# Generate 1000 data for the temperature difference of 5 years mean of 20 cities
c=[]
for n in range(1000):
    b=[]
    for x in range(20):
        array=np.array(city_random[x])
        LIST=array.tolist()[0]
        a1=np.mean(np.random.choice(LIST, 5, replace=True))
        a2=np.mean(np.random.choice(LIST, 5, replace=True))
        b.append(a1-a2)
    c.append(np.mean(b))
```

```python
len(c)
```

```python
# The distribution of the temperature difference of 5 years mean of 20 cities, if there is no climate change

np.percentile(c,[5,95])
```

```python
# We got 0.66 from the average difference between(1995-1999)and (2009-2014), 0.66>0.457, it is above 99% of the possible value. 
# No climate change is unlikely.
np.percentile(c,[1,99])
```
